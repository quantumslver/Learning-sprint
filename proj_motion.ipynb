```python
# Projectile motion.
# y(t) = ut + 1/2 at^2

import math
import matplotlib.pyplot as plt
import numpy as np


u = float(input("enter the value of initial velocity:"))
s = - 9.8

t = np.linspace(0, 1, 10)

y = u*t + 0.5*s*t**2

plt.plot(t,y, 'rx')

plt.xlabel("Time(s)")
plt.ylabel("Height(m)")
plt.xlim(0,2)
plt.ylim(0,2)

plt.title("Projectile Motion.")

plt.grid(True)
plt.show()

print(y[:5])
```
